```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
l = 633e-9
k = 2 * np.pi/l
x = 0
y = np.linspace(-30*l, 30*l, 200)
z = 10*l
r = np.sqrt(x**2 + y**2 + z**2)

k = 2 * np.pi / 633e-9
h = 1/(2*np.pi) * z/r * (1/r-1.0j*k) * np.exp(1.0j*k*r)/r
plt.plot(h.real, y)
plt.plot(np.abs(h), y)
```
