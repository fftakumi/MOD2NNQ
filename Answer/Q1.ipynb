# Reyleigh-Sommerfeldの回折回折公式を使った回折象の計算(点光源の図編)
## 前提知識
Reyleigh-Sommerfeldの回折公式
光の進行方向がを$z$方向である時、ある位置$(x,y,z)$での電場$U(x,y,z)$は
$$\begin{align*}
U(x,y,z) =\iint \dfrac{1}{2\pi} \dfrac{z}{r} \left( \dfrac{1}{r}-ik \right) \dfrac{exp(jkr)}{r}d\xi d\eta
\end{align*}
$$
で表される。ここで、$k$は波数、$\xi,\eta$は$z=0$平面における$x,y$座標であり、$r=\sqrt{(x-\xi)^2+(y-\eta)^2+z^2}$である。
この被積分関数$h(x,y,z)=\dfrac{1}{2\pi} \dfrac{z}{r} \left( \dfrac{1}{r}-ik \right) \dfrac{exp(jkr)}{r}$は点光源から放たれる球面波の式になっている。

## 目標
- Reyleigh-Sommerfeldの回折公式の被積分関数の形を知る。
$y=4\lambda$($y=0$にすると$r=0$になるときがあるから) における$x,z$平面をプロットする

## 条件
波長$\lambda$ : $633nm$
$-8\lambda \leq x \leq -8\lambda$
$y=4\lambda$
$-8\lambda \leq z \leq -8\lambda$

```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
l = 633e-9
k = 2 * np.pi/l
x = np.linspace(-8*l, 8*l, 200)
z = np.linspace(-8*l, 8*l, 200)
y = 4 * l

XX, ZZ = np.meshgrid(x, z)
R = np.sqrt(XX**2 + y**2 + ZZ**2)

k = 2 * np.pi / 633e-9
h = 1/(2*np.pi) * ZZ/R * (1/R-1.0j*k) * np.exp(1.0j*k*R)/R
plt.pcolormesh(ZZ, XX,  h.real, cmap="rocket")
plt.colorbar()
```
