```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
pix_size = 1.0e-6
width = 100
height = 100
z = 0.2e-3
wavelength = 633.0e-9
k = 2 * np.pi / wavelength

data = np.loadtxt("../Questions/data/5.csv")
```

```python
fft_image = np.fft.fft2(data)
u = np.fft.fftfreq(width, pix_size)
v = np.fft.fftfreq(height, pix_size)
UU, VV = np.meshgrid(u, v)
w = np.where(UU**2 + VV**2 <= 1/wavelength**2, np.sqrt(1/wavelength**2 - UU**2 - VV**2), 0)
h = np.exp(1.0j * 2*np.pi * w * z)
g = np.fft.ifft2(fft_image * h)
I = np.abs(g)**2/2
plt.imshow(I)
plt.title("0.2 mm")
plt.savefig("../Questions/images/Q5_2.png")
```
