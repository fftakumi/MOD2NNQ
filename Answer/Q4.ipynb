```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#(1)
t = np.linspace(0, 2*np.pi, 100)
omega = 1
Er_abs = 1

x_hat = np.array([1, 0]).reshape([-1,1])
y_hat = np.array([0, 1]).reshape([-1,1])
Er = Er_abs/np.sqrt(2) * (x_hat + 1.0j*y_hat)*np.exp(-1.0j*omega*t)
plt.figure(figsize=(5,5))
plt.plot(Er[0,:].real, Er[1,:].real)
plt.savefig("../Questions/images/Q4_1.png")
```

```python
#(2)
t = np.linspace(0, 2*np.pi, 100)
omega = 1
Er_abs = 1
El_abs = 1
delta_phi = np.arange(0, np.pi, np.pi/6)

plt.figure(figsize=(5,5))
for phi in delta_phi:
    x_hat = np.array([1, 0]).reshape([-1,1])
    y_hat = np.array([0, 1]).reshape([-1,1])
    Er = Er_abs/np.sqrt(2) * (x_hat + 1.0j*y_hat)*np.exp(-1.0j*omega*t)
    El = El_abs/np.sqrt(2) * (x_hat - 1.0j*y_hat)*np.exp(-1.0j*omega*(t-phi))
    E = Er + El
    plt.plot(E[0,:].real, E[1,:].real, label=r"$\Delta \phi=$" + str(phi))

plt.legend()
plt.savefig("../Questions/images/Q4_2.png")
```
