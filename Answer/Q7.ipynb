```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
z = np.linspace(0, 4*np.pi, 100)

phi = np.pi/6
eta =  0
Psi = phi + 1.j*eta
theta = np.pi/6 * 0

R = np.array([[np.cos(-theta), -np.sin(-theta)],
              [np.sin(-theta), np.cos(-theta)]])

r_hat = np.array([1,1.j]).reshape((-1, 1))
l_hat = np.array([1,-1.j]).reshape((-1, 1))

rcp = r_hat * np.exp(-1.j*z)*np.exp(1.j*Psi)
lcp = l_hat * np.exp(-1.j*z)*np.exp(-1.j*Psi)

total = rcp + lcp
total = np.tensordot(R.reshape([2,2,1]), total.reshape([2,1,-1]))

fig = plt.figure()
ax1 = fig.add_subplot(121,projection='3d')
ax1.plot(total[0,:].real,z , total[1,:].real)
ax1.set_xlim(-2,2)
ax1.set_zlim(-2,2)

ax2 = fig.add_subplot(122)
ax2.plot(total[0,:], total[1,:])
ax2.set_xlim(-2,2)
ax2.set_ylim(-2,2)
ax2.set_aspect("equal")
plt.show()
```
