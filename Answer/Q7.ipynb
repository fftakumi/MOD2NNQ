```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# (1)
phi = np.pi/4
phi_pol = np.linspace(0, 2*np.pi, 100)
T = np.array([[np.cos(-phi_pol)**2, np.sin(-2*phi_pol)/2],
              [np.sin(-2*phi_pol)/2, np.sin(-phi_pol)**2]])

r_hat = np.array([1,1.j]).reshape((-1, 1))
l_hat = np.array([1,-1.j]).reshape((-1, 1))

rcp = r_hat * np.exp(1.j*phi) / np.sqrt(2)
lcp = l_hat * np.exp(-1.j*phi) / np.sqrt(2)

Ein = rcp + lcp

Eout = []
for i in range(len(phi_pol)):
    Eout.append(T[:,:,i] @ Ein)

Eout = np.array(Eout)
Iout = np.abs(Eout[:,0,0])**2/2 + np.abs(Eout[:,1,0]**2/2)
plt.plot(phi_pol, Iout)
plt.xticks(np.arange(0, 2*np.pi+0.1, np.pi/2), ["0", r"$\pi/2$", r"$\pi$", r"$3\pi/2$", r"$2\pi$"])
plt.savefig("../Questions/images/Q7_1.png")
```

```python
# (2)
z = np.linspace(0, 2*np.pi, 100)
phi = np.pi/4
delta = np.pi/2

r_hat = np.array([1,1.j]).reshape((-1, 1))
l_hat = np.array([1,-1.j]).reshape((-1, 1))

rcp = r_hat * np.exp(1.j*z) * np.exp(1.j*phi) / np.sqrt(2)
lcp = l_hat * np.exp(1.j*z) * np.exp(-1.j*phi) / np.sqrt(2)

Ein = rcp + lcp

M = np.array([[np.exp(1.j*delta/2),  0],
              [0, np.exp(-1.j*delta/2)]])

Eout = M @ Ein

plt.figure(figsize=(5,5))
plt.plot(Eout[0,:].real, Eout[1,:].real)
plt.ylim(-1, 1)
plt.xlim(-1, 1)
plt.savefig("../Questions/images/Q7_2.png")
```
